```python
# DELETING COLUMNS WHICH DONT HAVE IMPORTANT INFO:

df.drop(columns=['source1','source2','source3','statecode','statepatientnumber','statuschangedate','contractedfromwhichpatientsuspected','estimatedonsetdate','patientnumber'],axis=1,inplace=True)
```

```python
df
```

```python
stat = df.currentstatus
```

```python
stat.dropna()
```

```python
total_active_case = stat.str.count('Hospitalized').sum()
print('Total Active Cases are: ',total_active_case)
```

```python
total_recovered_case = stat.str.count('Recovered').sum()
print('Total Recovered Cases are: ',total_recovered_case)
```

```python
total_deaths = stat.str.count('Deceased').sum()
print('Total Death Cases are: ',total_deaths)
```

```python
confirmed_case = total_active_case + total_recovered_case + total_deaths
print("Total Condirmed Cases are: ",confirmed_case)
```

```python
a = [total_active_case,total_recovered_case,total_deaths]
b = ['Active Cases','Recovered Cases','Deceased Cases']
explode = (0,0.1,1)
plt.pie(a, labels = b, colors=['blue','lightgreen','red'], explode = explode, shadow=False, autopct='%1.1f%%', startangle = 159)
plt.show()
```

```python
g = df.gender
g
```

```python
g.dropna()
```

```python
male = g.str.count('M').sum()
print("Male Affected: ",male)
```

```python
female = g.str.count('F').sum()
print("Female Affected: ",female)
```

```python
a = [male,female]
b = ['Male','Female']
explode = (0.1,0)
plt.pie(a, labels = b, colors=['Lightblue','Pink'], explode = explode, shadow=True, autopct='%1.1f%%', startangle = 60,radius=2)
plt.show()
```
